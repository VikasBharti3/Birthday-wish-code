```python
import matplotlib.pyplot as plt
import pandas as pd
import io
import numpy as np
import seaborn as sns

data=pd.read_csv("Country.csv")
data.head()
plt.scatter(data['Economy..GDP.per.Capita.'],data['Happiness.Score'])
plt.scatter(data['Freedom'],data['Happiness.Score'])

# comparision of GDP and Happiness using a composite plot. For better idea of how happniess and GDP are connected
plt.xlabel("Financial Freedom ")
plt.ylabel("Happiness Score")
plt.title("Finacial Freedom vs Happiness")
```

```python
#pandas library
import pandas as pd
#matplotlin library
import matplotlib.pyplot as plt
#read sample csv file
df1 = pd.read_csv('WorldGDP.csv')
#selecting required columns in the plot
pdf=pd.DataFrame(df1,columns=['Cost Index', 'Purchasing power Income'])
#set labels for title, x axis and y axis

# filter the number of countries
#including 100 countries
pdf = df1.iloc[0:150]
ax=pdf.plot(title='Cost index vs purchasing power')
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
#display graph
plt.show()
#filtering using iloc function
print("\n",df1.iloc[0:50])

# These suprising results. See that countries with higher purchasing power are also expansive.
# Data is quite accurate as you can see the table also. I also verified data online. 
# Data is not 100% accurate but still is comparable with real numbers
```
